# Data Dictionary

<b> 1- Pat-ID</b>: patient ID


<b>2- D.o.Birth</b>: Date of birth


<b>3- Exam Date</b>:  refers to the date on which an examination or test is scheduled to be taken.


<b>4- Exam Time</b>: refers to the specific time at which an examination or test is scheduled to take place.


<b>5- Exam Eye</b>: efer to an eye exam (left , right), which is a medical examination that assesses a person's vision and eye health.


<b>6 ,7- K1 F (D) , K2 F (D)</b>: "K1," "F," and "D" are abbreviations used in eye tests to describe measurements related to the curvature of the cornea of the eye. These measurements are used to determine the appropriate prescription for corrective lenses.
"K1" refers to the first reading of the corneal curvature, also known as the steepest reading.
"F" refers to the flattest reading of the corneal curvature.
"D"refers to the overall difference in curvature between the steepest and flattest readings.


<b>8- Axis F (Flat)</b>: In an eye test, "Axis F (Flat)" typically refers to a measurement of the astigmatism (common vision problem)correction needed for an individual's eyes.


<b>9- Astig F (D)</b>:In an eye test, "Astig F (D)" typically refers to the degree of astigmatism correction(vision problem) needed for an individual's eyes.


<b>10- Pachy Apex</b>: "Pachymetry Apex" refers to a measurement of the thickness of the cornea at its thinnest point.


<b>11- Pachy pupil</b> :Pachymetry is a test that measures the thickness of the cornea, as I explained in my previous answer. The pupil, on the other hand, is the black circular opening in the center of the iris that allows light to enter the eye (specific test).


<b>12- Pachy Min</b>:  refers to the minimum thickness of the cornea as measured by pachymetry.


<b>13,14- Pachy Min Pos X , Pachy Min Pos Y </b>: specialized measurement in eye test.


<b>15- KMax Sagittal Front (D) </b>:In an eye test, "KMax Sagittal Front (D)" refers to a measurement of the steepest curvature of the cornea on its front surface, expressed in diopters (D).

```python
corr_matrix = data.corr()
high_corr_features = set()

for col in corr_matrix.columns:
    corr_values = corr_matrix[col]
    high_corr = corr_values[corr_values >= 0.8].index.tolist()
    if len(high_corr) > 1:
        high_corr_features.update(high_corr)

print("Features with correlation coefficient >= 0.8:")
print(high_corr_features)

# K1 F (D):
#     K2 F (D):
#         Axis F (flat): x
#             Astig F (D):	x
#                 Pachy Apex:
#                        Pachy Pupil:
#                         Pachy Min:
#                              Pachy Min Pos X:	x
#                                 Pachy Min Pos Y:	x
#                                     KMax Sagittal Front (D)
```

```python
corr_matrix = data.corr()
neg_corr_features = set()

for col in corr_matrix.columns:
    corr_values = corr_matrix[col]
    neg_corr = corr_values[corr_values <= -0.8].index.tolist()
    if len(neg_corr) > 1:
        neg_corr_features.update(neg_corr)

print("Features with correlation coefficient <= -0.8:")
print(neg_corr_features)
```

## Features Extraction

```python
from datetime import datetime

# Assuming the date of birth column is named 'DateOfBirth'
data['D.o.Birth:'] = pd.to_datetime(data['D.o.Birth:'] ,errors='coerce')  # Convert the column to datetime format

current_date = datetime.now()  # Get the current date

# Calculate the age by subtracting the birth dates from the current date
data['Age'] = (current_date - data['D.o.Birth:']) // pd.Timedelta(days=365.25)

# Print the resulting DataFrame with the 'Age' column
print(data)
```
